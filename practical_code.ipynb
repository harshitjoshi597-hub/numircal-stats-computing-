### Regula Falsi (False Position) Method

Regula Falsi is a root-finding algorithm that combines features of the bisection method and the secant method. It works by repeatedly approximating the root of a continuous function within an interval `[a, b]` where `f(a)` and `f(b)` have opposite signs. It uses the secant line between `(a, f(a))` and `(b, f(b))` to estimate the new approximation `c`, ensuring that `f(a)` and `f(c)` (or `f(c)` and `f(b)`) have opposite signs to narrow down the interval.

```python
def regula_falsi(func, a, b, tol=1e-6, max_iter=100):
    """
    Implements the Regula Falsi (False Position) method to find a root of a function.

    Args:
        func (function): The function for which to find the root.
        a (float): The lower bound of the interval.
        b (float): The upper bound of the interval.
        tol (float, optional): The tolerance for the approximation. Defaults to 1e-6.
        max_iter (int, optional): The maximum number of iterations. Defaults to 100.

    Returns:
        float: The approximate root of the function.
        None: If the method fails to converge or initial conditions are not met.
    """

    fa = func(a)
    fb = func(b)

    if fa * fb >= 0:
        print("Error: Function has same signs at interval endpoints. Regula Falsi might not converge.")
        return None

    for i in range(max_iter):
        c = (a * fb - b * fa) / (fb - fa)
        fc = func(c)

        if abs(fc) < tol:
            print(f"Converged in {i+1} iterations.")
            return c

        if fa * fc < 0:
            b = c
            fb = fc
        else:
            a = c
            fa = fc

    print("Max iterations reached. No convergence within tolerance.")
    return c

# Example usage:
# Define a function for which to find the root
def f(x):
    return x**3 - 2*x - 5

# Find the root using Regula Falsi
root = regula_falsi(f, 2, 3)

if root is not None:
    print(f"The approximate root is: {root:.6f}")
    print(f"f({root:.6f}) = {f(root):.6f}")
```

### Gauss-Seidel Method

The Gauss-Seidel method is an iterative technique used to solve a system of linear equations. It is similar to the Jacobi method but uses updated values as soon as they are computed. This usually leads to faster convergence, but convergence is not guaranteed for all systems of equations. It is particularly effective for large systems where the matrix is diagonally dominant.

```python
import numpy as np

def gauss_seidel(A, b, x0, tol=1e-6, max_iter=100):
    """
    Solves a system of linear equations Ax = b using the Gauss-Seidel method.

    Args:
        A (numpy.array): The coefficient matrix.
        b (numpy.array): The constant vector.
        x0 (numpy.array): The initial guess for the solution vector.
        tol (float, optional): The tolerance for convergence. Defaults to 1e-6.
        max_iter (int, optional): The maximum number of iterations. Defaults to 100.

    Returns:
        numpy.array: The approximate solution vector x.
        None: If the method fails to converge.
    """

    n = len(b)
    x = x0.copy() # Make a copy to avoid modifying x0 directly

    for k in range(max_iter):
        x_new = x.copy()
        for i in range(n):
            sum1 = np.dot(A[i, :i], x_new[:i])
            sum2 = np.dot(A[i, i+1:], x[i+1:])
            x_new[i] = (b[i] - sum1 - sum2) / A[i, i]

        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            print(f"Converged in {k+1} iterations.")
            return x_new

        x = x_new

    print("Max iterations reached. No convergence within tolerance.")
    return x

# Example usage:
# Define the coefficient matrix A and the constant vector b
A = np.array([
    [10., -1., 2., 0.],
    [-1., 11., -1., 3.],
    [2., -1., 10., -1.],
    [0., 3., -1., 8.]
])
b = np.array([6., 25., -11., 15.])

# Initial guess
x0 = np.zeros_like(b)

# Solve using Gauss-Seidel
solution = gauss_seidel(A, b, x0)

if solution is not None:
    print("\nApproximate solution x:")
    print(solution)
    print("\nCheck Ax = b:")
    print(np.dot(A, solution))
```
